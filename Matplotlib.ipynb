```python
plt.figure (figsize=(5,6)) #using scatter plot
plt.scatter(new_df['Max Price (Lakh)'],new_df['Insurance'],color='red')
plt.title('Max Price(Lakh) Vs Insurance')
plt.xlabel('Max Price (Lakh)')
plt.ylabel('Insurance')
plt.legend()
plt.grid (True)
plt.show()
```

```python
#drawing a regression line to detect outliers??
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.figure(figsize=(5, 6))  # Set figure size
plt.scatter(new_df['Max Price (Lakh)'], new_df['Insurance'], color='red', label='Data Points')
m, b = np.polyfit(new_df['Max Price (Lakh)'], new_df['Insurance'], 1)  # Linear fit: y = mx + b
plt.plot(new_df['Max Price (Lakh)'], m * new_df['Max Price (Lakh)'] + b, color='blue', label='Trend Line')

# Add titles and labels
plt.title('Max Price (Lakh) vs Insurance')
plt.xlabel('Max Price (Lakh)')
plt.ylabel('Insurance')
```

```python
#aggregate data by car model
sales_data = df.groupby('Category')['Sales'].sum()
sales_data
```

```python
#creating pie chart
plt.figure(figsize=(8,8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%')
plt.title('Proportion of Sales by Category')
plt.show()
```

```python
#creating a histogram
plt.hist(df['Sales'], bins=10, color ='green', edgecolor ='violet')
plt.show()
```

```python
plt.title('Proportion of Sales by Category')
plt.xlabel('bins')
plt.ylabel('frequency')
plt.hist(df['Sales'], bins=10, color ='green', edgecolor ='violet')
plt.show() #helps in identifying frequency ranges
```

```python
#creating an area plot, which is similar to line plot
plt.fill_between(df['Date'], df['Sales'], color='white', alpha=0.3)
plt.plot(df['Date'], df['Sales'], color='Violet') #line plot
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45) #used to rotate the 
plt.show()
```

```python
plt.fill_between(df['Date'], df['Sales'], color='white', alpha=0.20)
plt.plot(df['Date'], df['Sales'], color='Violet') #line plot
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45) #used to rotate the 
plt.show()
```
