```python
car = pd.read_excel('car/Car data.xlsx')
car.head()
```

```python
car.columns #to enable easier copy pasting and working fast
```

```python
variables = ['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC','Seats','Variants','Type', 'Ex-Showroom Price',	'RTO',	'Insurance','Other','Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i, var in enumerate(variables,start=1):
    plt.subplot(4, 3, i) #specify how we want to shape them
    sns.scatterplot(x=car[var], y=car['Max Price (Lakh)'], alpha=0.7)
    plt.title(f'Scatter plot of {var} vs Onroad Price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i, var in enumerate(variables,start=1):
    plt.subplot(3, 4, i) #specify how we want to shape them
    sns.histplot(x=car[var], bins=20, color='Yellow', edgecolor='Pink', kde = True)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#create a boxplot of Range by Car Typ)
colors = ['yellow', 'blue']
sns.boxplot(data=cars, x='Type', y='Range (kmpl)',hue = 'Type', palette=colors)
plt.title('Boxplot of Range by Car Type')
plt.show()
```

```python
selected_columns = ['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC','Seats', 'Variants', 'Ex-Showroom Price', 'RTO', 'Insurance',
       'Other']
df_selected= cars[selected_columns]
df_selected.head()
```

```python
#finding correlation??
corr= df_selected.corr()
corr
```

```python
#importance of correlation?needs more studies
sns.heatmap(corr, cmap='Reds')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
#used to add text or markers to specific locations on a plot to provide additional information or highlight certain data points.
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
#violin plot of max price by car type 
sns.violinplot(data=cars, x='Type', y='Max Price (Lakh)')
plt.title('Violin Plot of Maximum Price by Car Type')
plt.show()
```
