```python
q1 = df['Sales'].quantile(0.25) #getting the quartile range in q1 and q3
q3 = df['Sales'].quantile(0.75)
iqr = q3-q1
print(q1, q3, iqr)
```

```python
lower_bound=q1-1.5*iqr
```

```python
lower_bound
```

```python
upper_bound=q3+1.5*iqr
```

```python
upper_bound
```

```python
required=df[(df['Sales'] >= lower_bound) & (df['Sales'] <= upper_bound)]
print("All Sales\n", df["Sales"])
print("Sales after removing outliers\n",required["Sales"].values)      
```
